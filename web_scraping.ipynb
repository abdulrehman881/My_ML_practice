```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# headers = {'User-Agent':'Mozilla/5.0(Windows NT 6.3;Win64; x64)AppleWebkit/537.36
webpage=requests.get('https://moviebox.ph/').text
```

```python
soup = BeautifulSoup(webpage, 'html.parser')
```

```python
soup.find_all('h1')[0].text
```

```python
for i in soup.find_all('h2'):
    print(i.text)
```

```python
for i in soup.find_all('p',class_="w-full text-white/80 font-normal text-[12px] leading-[14px] truncate md:text-[14px] md:leading-[15px]"):
    print(i.text)
```
